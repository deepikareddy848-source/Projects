## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Handling missing values
df['Item_price']=df['Item_price'].fillna(df['Item_price'].median())
df['Customer Remarks'] = df['Customer Remarks'].fillna("No Remarks")
df['Customer_City']=df['Customer_City'].fillna("Other")
````

````python
df['Issue_reported at'] = pd.to_datetime(df['Issue_reported at'], format='%d/%m/%Y %H:%M')
df['issue_responded'] = pd.to_datetime(df['issue_responded'], format='%d/%m/%Y %H:%M')

df['Issue_reported at(formatted)']=df['Issue_reported at'].dt.strftime('%d%m%y')
df['issue_responded (formatted)']=df['issue_responded'].dt.strftime('%d%m%y')

df['Time difference']=(df['issue_responded']-df['Issue_reported at']).dt.total_seconds()/6
df.head(3)
````

````python
if 'connected_handling_time' in df.columns:
  df.drop(columns=['connected_handling_time'],inplace=True)

if 'order_date_time' in df.columns:
  df.drop(columns=['order_date_time'],inplace=True)



````

````python
rem_columns=['Order_id','Product_category']
exist_columns_remove=[col for col in rem_columns if col in df.columns]
if exist_columns_remove:
  df=df.drop(columns=exist_columns_remove)
  print('Columns removed successfully')
else:
  print('No columns to remove')
````

#### Chart - 7

````python
# Chart - 7 visualization code
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convert 'Issue_reported at' and 'issue_responded' to datetime objects
df['Issue_reported at'] = pd.to_datetime(df['Issue_reported at'], format='%d/%m/%Y %H:%M', errors='coerce')
df['issue_responded'] = pd.to_datetime(df['issue_responded'], format='%d/%m/%Y %H:%M', errors='coerce')

# Extract the date
df['Issue_reported_date'] = df['Issue_reported at'].dt.date
df['Issue_responded_date'] = df['issue_responded'].dt.date

# Group by date and count the number of issues reported and responded
reported_counts = df.groupby('Issue_reported_date').size()
responded_counts = df.groupby('Issue_responded_date').size()

# Create a DataFrame for plotting
time_series_data = pd.DataFrame({'Reported': reported_counts, 'Responded': responded_counts}).fillna(0)

# Plotting the line chart
plt.figure(figsize=(15, 7))
sns.lineplot(data=time_series_data)
plt.title('Number of Issues Reported and Responded Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A line chart is suitable for visualizing trends over time. By plotting the number of issues reported and responded to on a daily basis, we can observe patterns, peaks, and dips in customer service activity.

##### 2. What is/are the insight(s) found from the chart?

This chart will show the volume of issues reported and responded to each day. It can help identify days with high customer service traffic, potential backlogs (if reported issues consistently exceed responded issues), or periods of low activity.

````python
# Chart - 9 visualization code
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the average CSAT score for each combination of channel and agent shift
average_csat_by_channel_shift = df.groupby(['channel_name', 'Agent Shift'])['CSAT Score'].mean().unstack()

plt.figure(figsize=(10, 6))
sns.heatmap(average_csat_by_channel_shift, annot=True, cmap='viridis', fmt=".2f")
plt.title('Average CSAT Score by Channel and Agent Shift')
plt.xlabel('Agent Shift')
plt.ylabel('Channel Name')
````

#### Chart - 10

````python
# Chart - 10 visualization code
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the average CSAT score for each combination of Tenure Bucket and Agent Shift
average_csat_by_tenure_shift = df.groupby(['Tenure Bucket', 'Agent Shift'])['CSAT Score'].mean().unstack()

# Plotting the grouped bar chart
average_csat_by_tenure_shift.plot(kind='bar', figsize=(12, 7))
plt.title('Average CSAT Score by Tenure Bucket and Agent Shift')
plt.xlabel('Tenure Bucket')
plt.ylabel('Average CSAT Score')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Agent Shift', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A grouped bar chart is suitable for comparing a numerical value 'Average CSAT Score' across combinations of two categorical variables 'Tenure Bucket' and 'Agent Shift'. It allows for easy comparison of average CSAT within each tenure bucket across different shifts, and vice versa.

##### 2. What is/are the insight(s) found from the chart?

This chart will show how the average CSAT score is influenced by both the agent's tenure and their shift. It can reveal if certain tenure groups perform better in specific shifts, or if there's a consistent pattern across all shifts for a particular tenure bucket.
